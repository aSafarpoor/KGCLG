```python
catedges = [[],[]]
for i in tqdm(range(len(meta))):
    row = meta.iloc[i]
    cats = row.category
    # catdi = catd[i]
    catasin = row.asin
    for cat in cats:
        catedges[0].append(catasin)
        catedges[1].append(cat)
```

```python
z = np.zeros(11)
for i in tqdm(range(len(meta))):
    row1 = meta.iloc[i].brand 
    z[min(10,len(row1))]+=1
```

```python
for i in z:
    print(i)
```
